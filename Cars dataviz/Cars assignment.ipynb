```python
sns.boxplot(data = cars, x="model_year", y="mpg")  # creating boxplot for model_year and mpg
plt.savefig("images/boxplot_model_year_mpg.png")
```

```python
# using statsmodels for statistical analysis using Ordinary Least Squares. This is a 
# bivariate linear regression for the x-variable model_year and the y-variable mpg.
# We can see several things in the table:
# R-squared = 0.337 means that the variation in the x-variable accounts for 33.7 percent
# of the variation in the y-variable
# Prob (F-statistic) = 1.08e-36 means that the chance that this correlation appeared out of chance is
# infintesmal (t-value 14.080)
# coef = 1.23 means that by average the mpg-value increase by 1.23 for each year.

results = smf.ols('mpg ~ model_year', data=cars).fit()
results.summary()
```

```python
# We note that as time goes by the mpg increase.
# We also note that the amount of horsepower decrease and that a decrease in horsepower correlates
# with an increase in mpg.
# Could that be the explanation for the increase in mpg over time?

sns.lmplot(data=cars, x="model_year", y="mpg")
plt.savefig("images/model_year_mpg.png")
sns.lmplot(data=cars, x="horsepower", y="mpg")
plt.savefig("images/horsepower_mpg.png")
sns.lmplot(data=cars, x="model_year", y="horsepower")
plt.savefig("images/model_year_horsepower.png")
```

```python
# Doing a multivariate regression analysis we find that the correlation between model_year and mpg
# stands. Even after we control for the effect of decreasing horsepower we note that we still a very high t-value 
# (dropped from 14.080 to 9.919).
# The coefficient is still positive, although smaller (from 1.23 to 0.6573)

results = smf.ols("mpg ~ model_year + horsepower", data=cars).fit()
results.summary()
```
