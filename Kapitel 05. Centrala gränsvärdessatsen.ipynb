<h1>Centrala gränsvärdessatsen utvärderad för olika fördelningar</h1>

```python
# Importera nödvändiga paket
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3>Vi börjar med att utvärdera CGS för normalfördelningen. Resten av notebooken kommer använda exakt samma metodik, fast byta ut fördelningen.</h3>  
1. Importera normalfördelningen från SciPy. <br>
2. Skapa: en variabel 'iter', för antalet iterationer. En NumPy array 'means', fylld med nollor, lika stor som antalet iterationer. Här tänker vi spara våra beräknade medelvärden. Och en lista 'n_list', för våra olika samplingsnivåer n. <br>
3. Skapa en tom figur 'fig' med matplotlib subplots(). <br> 
4. Loopa genom alla samplingsnivåer i n_list. <br> 
5. Loopa över antalet iterationer, och sampla ur den valda fördelningen med .rvs(). Sätt size=n för att hämta samplingsnivån. Räkna ut medelvärdet på stickprovet med .mean(), och spara värdet i vår array 'means'. <br> 
6. Plotta fördelningen av medelvärden med Seaborns .kdeplot(). <br> 
7. Lägg till en legend i figuren med .legend() och sätt titeln på legenden.

```python
# 1
from scipy.stats import norm # Importera normalfördelningen från SciPy

#2 
iter = 5000 # Antalet iterationer
means = np.zeros(iter) # Skapa en tom NumPy-array med plats för alla våra medelvärden
n_list = [2,3,5,20,50] # Skapa en lista med värden på n

#3
fig = plt.subplots() # Skapa tom figur

#4
for n in n_list: # Loopa över alla värden på n
    
    #5
    for i in range(iter): # Den här loopen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = norm.rvs(size=n) # Tag n slumpmässiga värden från normalfördelningen med norm.rvs()
        mean = rv.mean() # Bilda medelvärdet av dem
        means[i] = mean # Spara medelvärdet i listan means
    
    #6
    fig = sns.kdeplot(means) # Plotta medelvärdena med en KDE-plot i Seaborn


#7
fig.legend(n_list).set_title('n (# samples)') # Lägg till legend med värde på 'n'
```

Vi ser här att redan för n=2 ser medelvärdet för stickprovet ut att vara normalfördelat. Det gäller för normalfördelningen att om en slumpvariabel X är normalfördelad, så kommer medelvärdet av X också alltid vara normalfördelat.

<h3>Exponentialfördelningen</h3>
Vi följer exakt samma metodik som för normalfördelningen, men byter ut fördelningen i .rvs() till exponentialfördelningen.

```python
# 1
from scipy.stats import expon # Importera exponentialfördelningen från SciPy

#2 
iter = 5000 # Antalet iterationer
means = np.zeros(iter) # Skapa en tom NumPy-array med plats för alla våra medelvärden
n_list = [2,3,5,20,50] # Skapa en lista med värden på n

#3
fig = plt.subplots() # Skapa tom figur

#4
for n in n_list: # Loopa över alla värden på n
    
    #5
    for i in range(iter): # Den här loopen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = expon.rvs(size=n) # Tag n slumpmässiga värden från exponentialfördelningen med expon.rvs()
        mean = rv.mean() # Bilda medelvärdet av dem
        means[i] = mean # Spara medelvärdet i listan means
    
    #6
    fig = sns.kdeplot(means) # Plotta medelvärdena med en KDE-plot i Seaborn

#7
fig.legend(n_list).set_title('n (# samples)') # Lägg till legend med värde på 'n'
```

Här är det mer tydligt att för låga n följer medelvärdet av vårt stickprov fortfarande i viss mån sin ursprungsfördelning (exponentialfördelningen). 
När n blir större, blir också medelvärdet mer och mer normalfördelat, tills det vid n=50 ser ut att mer eller mindre exakt följa en normalfördelad slumpvariabel.

<h3>Diskreta fördelningar - Binomialfördelningen</h3>

Vi kikar nu på binomialfördelningen, för att se hur CGS fungerar för en diskret variabel. Metodiken är återigen samma som innan.

```python
# 1
from scipy.stats import binom # Importera binomialfördelningen från SciPy

#2 
iter = 5000 # Antalet iterationer
means = np.zeros(iter) # Skapa en tom NumPy-array med plats för alla våra medelvärden
n_list = [2,3,5,20,50] # Skapa en lista med värden på n

#3
fig = plt.subplots() # Skapa tom figur

#4
for n in n_list: # Loopa över alla värden på n
    
    #5
    for i in range(iter): # Den här loopen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = binom.rvs(n=5, p=0.5, size=n) # Tag n slumpmässiga värden från binomialfördelningen med binom.rvs()
        mean = rv.mean() # Bilda medelvärdet av dem
        means[i] = mean # Spara medelvärdet i listan means
    
    #6
    fig = sns.kdeplot(means) # Plotta medelvärdena med en KDE-plot i Seaborn

#7
fig.legend(n_list).set_title('n (# samples)') # Lägg till legend med värde på 'n'
```

För låga n är fördelningen fortfarande väldigt svajig, men när n går upp mot 20-50 konvergerar resultatet mot att medelvärdet följer normalfördelningen.

```python
# 1
from scipy.stats import uniform # Importera likformiga fördelningen från SciPy

#2 
iter = 5000 # Antalet iterationer
means = np.zeros(iter) # Skapa en tom NumPy-array med plats för alla våra medelvärden
n_list = [2,3,5,20,50] # Skapa en lista med värden på n

#3
fig = plt.subplots() # Skapa tom figur

#4
for n in n_list: # Loopa över alla värden på n
    
    #5
    for i in range(iter): # Den här loopen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = uniform.rvs(size=n) # Tag n slumpmässiga värden från likformiga fördelningen med uniform.rvs()
        mean = rv.mean() # Bilda medelvärdet av dem
        means[i] = mean # Spara medelvärdet i listan means
    
    #6
    fig = sns.kdeplot(means) # Plotta medelvärdena med en KDE-plot i Seaborn

#7
fig.legend(n_list).set_title('n (# samples)') # Lägg till legend med värde på 'n'
```

Även här fungerar CGS bra. Vid mycket låga n får medelvärdet av den likformiga fördelningen ett triangelutseende. För höga n konvergerar de mot en normalfördelning.

<h3>Fungerar alltid centrala gränsvärdessatsen?</h3>
Det enklar svaret är nej. CGS håller inte alltid för alla fördelningar. Det finns fördelningar där medelvärdet konvergerar mot andra fördelningar istället, och det finns fördelningar som inte
har något definierat medelvärde. Sådana fördelningar kallas ibland på engelska 'pathological distributions'. Om vi räknar ut medelvärdet för ett stickprov från en sådan fördelning kommer vi se att
oavsett hur många samplenivåer vi använder oss av så konvergerar aldrig medelvärdet till en normalfördelning.

<h3>t-fördelningen med en frihetsgrad</h3>
Vi exemplifierar problemet genom att testa för t-fördelningen med en frihetsgrad. Vi kommer att kika på t-fördelningen noggrannare om ett tag när vi kommer till konfidensintervall ochg hypotestestning.
t-fördelningen med en frihetsgrad är ett specialfall av en fördelning som kallas för Cauchyfördelningen. 
Cauchyfördelningen har inget definierat medelvärde (och inte heller någon definierad varians). Det här kan vara förvånande eftersom Cauchyfördelningen till första anblicken är mycket lik
normalfördelningen.

```python
from scipy.stats import t # Importera t-fördelningen från SciPy

iter = 5000 # Antalet iterationer
means = np.zeros(iter) # Skapa en tom NumPy-array med plats för alla våra medelvärden
n_list = [2,3,5,20,50] # Skapa en lista med värden på n

fig, ax = plt.subplots() # Skapa tom figur

for n in n_list: # Loopa över en lista med olika värden på n
    
    #5
    for i in range(iter): # Den här loopen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = t.rvs(df=1, size=n) # Tag n slumpmässiga värden från t-fördelningen (1 frihetsgrad) med t.rvs()
        mean = rv.mean() # Bilda medelvärdet av dem
        means[i] = mean # Spara medelvärdet i listan means
        
    #6
    fig = sns.kdeplot(means) # Plotta medelvärdena med en KDE-plot i Seaborn
#7
ax = ax.set(xlim=(-100,100)) # Sätt gränsvärden för x-axel
fig.legend(n_list).set_title('n (# samples)') # Lägg till legend med värde på 'n'
```

Det blir tydligt här att fördelningen inte konvergerar någonstans. Beteendet för medelvärdena verkar närmast kaotiskt. Cauchy-fördelningen är alltså en fördelning för vilken centrala gränsvärdessatsen <i>inte gäller</i>.
